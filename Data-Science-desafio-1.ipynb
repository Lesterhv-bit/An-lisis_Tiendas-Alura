### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



*   ***Para este análisis se realiza la sumatoria total de los precios de cada uno de los articulos obteniendo como resultado la facturación total por cada Tienda.***






```python
ingresos_tienda_1 = tienda["Precio"].sum()
ingresos_tienda_2 = tienda2["Precio"].sum()
ingresos_tienda_3 = tienda3["Precio"].sum()
ingresos_tienda_4 = tienda4["Precio"].sum()
print(f"Facturación T1 {ingresos_tienda_1}")
print(f"Facturación T2 {ingresos_tienda_1}")
print(f"Facturación T3 {ingresos_tienda_1}")
print(f"Facturación T4 {ingresos_tienda_1}")
```

```python
import matplotlib.pyplot as plt
facturacion_total = ingresos_tienda_1 + ingresos_tienda_2 + ingresos_tienda_3 + ingresos_tienda_4
print(f"El total de ingresos de todas las tiendas fue de:$ {facturacion_total}")

plt.bar (x=["T1","T2","T3","T4"] ,height=(ingresos_tienda_1, ingresos_tienda_2, ingresos_tienda_3, ingresos_tienda_4), color ="red")
plt.title("Total de Ingresos por Tienda")
plt.ylabel("Ingresos")
plt.xlabel("Tiendas")
plt.show()
```

# 2. Ventas por categoría


*   ***Agrupando los productos segun la columna de "Categoría del Producto" usando tienda.grouphy, sumando valores de precio dentro de cada grupo y mediante .sort_values("Precio, ascending= False") los ordeno los resultados de manera descendente.***



```python
datos_agrupados_1 = tienda.groupby(["Categoría del Producto"])[["Precio"]].sum().sort_values("Precio",ascending=False)
datos_agrupados_1
```

En la Tienda 1 la mayor venta es en la categoría de "Electrónicos" y la que menos vende es "Libros"

```python
datos_agrupados_2 = tienda2.groupby(["Categoría del Producto"])[["Precio"]].sum().sort_values("Precio",ascending=False)
datos_agrupados_2
```

En la Tienda 1 la mayor venta es en la categoría de "Electrónicos" y la que menos vende es "Libros"

```python
datos_agrupados_3 = tienda3.groupby(["Categoría del Producto"])[["Precio"]].sum().sort_values("Precio",ascending=False)
datos_agrupados_3
```

En la Tienda 1 la mayor venta es en la categoría de "Electrónicos" y la que menos vende es "Libros"

```python
datos_agrupados_4 = tienda4.groupby(["Categoría del Producto"])[["Precio"]].sum().sort_values("Precio",ascending=False)
datos_agrupados_4
```

En la Tienda 1 la mayor venta es en la categoría de "Electrónicos" y la que menos vende es "Libros"

Resumiendo se puede apreciar que en todas las tiendas las mayores vetnas proviene de la categoría "Electrónicos" mientras que que la categoría "Libros" es la que menos vende.

Gráficando los resultados de ventas según las categorías por cada tienda

```python
import matplotlib.pyplot as plt
datos_agrupados_1.plot.bar(color="red")
plt.title("Ventas por categoría en Tienda 1")
plt.ylabel("Ventas")
plt.xlabel("Categoría")
plt.grid(axis="y")
plt.show()


datos_agrupados_2.plot.bar(color="red")
plt.title("Ventas por categoría en Tienda 2")
plt.ylabel("Ventas")
plt.xlabel("Categoría")
plt.grid(axis="y")
plt.show()


datos_agrupados_3.plot.bar(color="red")
datos_agrupados_4.plot.bar(color="red")
plt.title("Ventas por categoría en Tienda 3")
plt.ylabel("Ventas")
plt.xlabel("Categoría")
plt.grid(axis="y")
plt.show()

datos_agrupados_4.plot.bar(color="red")
plt.title("Ventas por categoría en Tienda 4")
plt.ylabel("Ventas")
plt.xlabel("Categoría")
plt.grid(axis="y")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Para la tienda 1
calificacion_tienda_1 = tienda["Calificación"].mean().round(7)
print(f"La Calificación de la Tienda 1 es de: {calificacion_tienda_1}")

# Para la tienda 2
calificacion_tienda_2 = tienda2["Calificación"].mean().round(7)
print(f"La Calificaciónn de la Tienda 2 es de: {calificacion_tienda_2}")

# Para la tienda 3
calificacion_tienda_3 = tienda3["Calificación"].mean().round(7)
print(f"La Calificaciónde la Tienda 3 es de: {calificacion_tienda_3}")

# Para la tienda 3
calificacion_tienda_4 = tienda4["Calificación"].mean().round(7)
print(f"La Calificación de la Tienda 4 es de: {calificacion_tienda_4}")
```

```python
promedio_calificación_tienda = []
promedio_calificación_tienda.append(calificacion_tienda_1)
promedio_calificación_tienda.append(calificacion_tienda_2)
promedio_calificación_tienda.append(calificacion_tienda_3)
promedio_calificación_tienda.append(calificacion_tienda_4)
print(promedio_calificación_tienda)

promedio_calificación_tienda
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= ["T1","T2","T3","T4"]
y = promedio_calificación_tienda
plt.title("Calificaciones promedio de Tiendas")

plt.scatter(x,y,color="red", s= 50)
plt.ylim(3.9,4.09)
plt.yticks(np.arange(3.9, 4.09, 0.01))
plt.tight_layout()
plt.grid(axis="y")

plt.show()
```
